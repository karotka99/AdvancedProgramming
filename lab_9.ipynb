```python
img4_grey = cv2.imread('4.jpg',0)
img4_matrix1=cv2.threshold(cv2.GaussianBlur(img4_grey, (5, 5), 0), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
img4_matrix2=cv2.threshold(cv2.bilateralFilter(img4_grey, 5, 75, 75), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
img4_matrix3=cv2.threshold(cv2.medianBlur(img4_grey, 3), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
img4_matrix4=cv2.adaptiveThreshold(cv2.GaussianBlur(img4_grey, (5, 5), 0), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
img4_matrix5=cv2.adaptiveThreshold(cv2.bilateralFilter(img4_grey, 9, 75, 75), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
img4_matrix6=cv2.adaptiveThreshold(cv2.medianBlur(img4_grey, 3), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
cv2.imwrite("image4_gray1.jpg", img4_matrix1)
cv2.imwrite("image4_gray2.jpg", img4_matrix2)
cv2.imwrite("image4_gray3.jpg", img4_matrix3)
cv2.imwrite("image4_gray4.jpg", img4_matrix4)
cv2.imwrite("image4_gray5.jpg", img4_matrix5)
cv2.imwrite("image4_gray6.jpg", img4_matrix6)
```
