### 1. Write a function to encrypt the contents of a file using Caesar cipher. Use the ord() and chr() methods, which return the numeric value of a character and convert a number to a character, respectively.

```python
f=open("tekstowy.txt","r")
plain_text=f.read()
f.close()


def encryption(text):
    ciphertext=""
    for i in text:
        ciphertext+=chr((ord(i)+3)%256)
    return ciphertext
ciphertext=encryption(plain_text)

f=open("tekst_zaszyfrowany.txt","w")
f.write(ciphertext)
f.close()
```

```python
f=open("tekst_zaszyfrowany.txt","r")
plain_text=f.read()
f.close()
def decryption(text):
    ciphertext=""
    for i in text:
        ciphertext+=chr((ord(i)-3)%256)
    return ciphertext
descriptor = decryption(plain_text)

f=open("tekst_odszyfrowany.txt","w")
f.write(descriptor)
f.close()
```

### 2. Write a function that generates a graphic showing the approximation of the value of the number Ï€ by the Monte-Carlo method.

```python
import random

def pi_approx(n):
    how_many=0
    for i in range(n):
        x=random.random()
        y=random.random()
        if x**2+y**2<1:
            how_many+=1
    return how_many/n*4

pi_approx(10**6)
```

```python
import matplotlib.pyplot as plt

def plot_pi_approx(n):
    plt.figure(figsize=(6,6))
    for i in range(n):
        x=random.random()
        y=random.random()
        if x**2+y**2<1:
            plt.plot(x,y,".r")
        else:
            plt.plot(x,y,".b")
    plt.show()        

plot_pi_approx(10**4)    
```

### 4. Using the %timeit function, check which of the procedures is faster: a list with sines whose arguments are consecutive integers (in the range $0 - 10^6$ ) created using the functions math.sin and np.sin

```python
import math

tabl1=[math.sin(i) for i in range(10**6)]
tabl2=np.sin(np.arange(10**6))
print(tabl1)
print(tabl2)
```

```python
import math

%timeit -n 1 -r 3 [math.sin(i) for i in range(10**6)]
%timeit -n 1 -r 3 np.sin(np.arange(10**6))
```

### 6. Write a function that generates a matrix with random elements (integers in a given range) with a predetermined determinant value.

```python
import numpy.linalg as lg

def random_matrix(a,b,det):
    matrix = np.random.randint(a,b,(2,2))
    while lg.det(matrix) != det:
        matrix=np.random.randint(a,b,(2,2))
    return matrix

random_matrix(-10,10,2)
```
