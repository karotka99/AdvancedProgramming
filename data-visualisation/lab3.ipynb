### 1. Write a procedure that displays the graph of the sinus and cosinus functions with axis modification, captions, legend, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi

x = np.linspace(0,2*pi,100)
y = [np.sin(x), np.cos(x)]

plt.plot(x, y[0], label="$\sin x$")
plt.plot(x,y[1], label="$\cos x$")


plt.xlabel("$x$", fontsize=15)
plt.ylabel("$y$", fontsize=15)


plt.xlim(0,2*pi)
plt.grid()
plt.legend(loc="best")
plt.title("Sinus and cosinus plot")
plt.xticks([0,pi/2,pi,3*pi/2,2*pi],["0",r"$\frac{\pi}{2}$","$\pi$",r"$\frac{3\pi}{2}$", "$2\pi$"])
plt.yticks([-1,-0.5,0,0.5,1],[-1,-0.5,0,0.5,1])

plt.savefig("sin_cos.png")

plt.tight_layout()
plt.show()


```

### 2. Write a procedure that displays a random number (position, color and type) of geometric figures.

```python
import random

def shuffle(n):
    return random.random()*n

marker=["o","d","p","*", "<","^","s","h","D/d"]
for i in range(20):
 x= random.random() 
 y=random.random()
 plt.plot(x, y,marker[int(shuffle(6))],ms=shuffle(50))



plt.show()
```

### 3. Write a procedure that displays subplots with the values of the function $f(x)={x,x^2,x^3,x^4}$.

```python
x = np.linspace(-2,2,100)
y=[x,x**2,x**3,x**4]
labels = ["$x$","$x^2$","$x^3$","$x^4$"]


for i in range(4):
 plt.subplot(2,2,i+1)
 plt.plot(x,y[i])
 plt.grid()
 plt.xlabel("$x$", fontsize=15)
 plt.ylabel(labels[i], fontsize=15)
 

plt.tight_layout()
plt.show()
```

```python
x=np.linspace(-2,2,100)
y=[x,x**2,x**3,x**4]
labels = ["$x$","$x^2$","$x^3$","$x^4$"]

fig, axs = plt.subplots(2,2)

axs=axs.ravel()

for i in range(len(axs)):
    axs[i].plot(x,y[i])
    axs[i].set_xlabel("$x$",fontsize=15)
    axs[i].set_ylabel(labels[i],fontsize=15)
    axs[i].grid()
    
    
```
