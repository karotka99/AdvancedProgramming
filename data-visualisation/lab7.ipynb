  + zestawienie nauczycieli akademickich zatrudnionych na dowolnym etacie w latach 2014-2018 z podziałem na województwo
  + zestawienie różnych stanowisk (asystenci, adiunkcji, profesorowie) zatrudnionych w latach 2014-2018 w województwie śląskim
  + zestawienie nauczycieli akademickich zatrudnionych na uniwersytetach i szkołach technicznych z podziałem na płcie w województwie śląskim
1. W danych znajdź:
  + wartości dla województwa śląskiego
  + województwa, w których jest zatrudnionych najwięcej i najmniej nauczycieli akademickich
  + województwa, które zatrudniają kobiety ponad średnią


```python
import pandas as pd
```

```python
dane = pd.read_csv("nauczyciele.csv", sep=";")
dane.head()
```

```python
del dane[dane.columns[-1]]
del dane[dane.columns[0]]
dane.head()
```

```python
_dane = []
for i in dane.values:
    temp = []
    for j in i:
        temp.append(j.replace(',','.'))
    _dane.append(temp)
            
        
df = pd.DataFrame(_dane, dtype="float64", columns=dane.columns, index=dane['Nazwa'])
del df['Nazwa']
df.head()
```

## Wykresy

```python
import matplotlib.pyplot as plt

dane_14 = df["nauczyciele akademiccy;ogółem;uniwersytety;2014;[osoba]"][1:]
dane_15 = df["nauczyciele akademiccy;ogółem;uniwersytety;2015;[osoba]"][1:]
dane_16 = df["nauczyciele akademiccy;ogółem;uniwersytety;2016;[osoba]"][1:]
dane_17 = df["nauczyciele akademiccy;ogółem;uniwersytety;2017;[osoba]"][1:]
dane_18 = df["nauczyciele akademiccy;ogółem;uniwersytety;2018;[osoba]"][1:]
woj = df.index[1:]

plt.plot(range(len(dane_14)), dane_14, "ro", label="2014")
plt.plot(range(len(dane_15)), dane_15, "bo", label="2015")
plt.plot(range(len(dane_16)), dane_16, "yo", label="2016")
plt.plot(range(len(dane_17)), dane_17, "ko", label="2017")
plt.plot(range(len(dane_18)), dane_18, "go", label="2018")
plt.xticks(range(len(woj)), woj, rotation=70)
plt.title('Nauczyciele akademiscy zatrudnieni w latach 2014-2018')
plt.ylabel('osoby', fontsize=15)
plt.legend(loc='best')

plt.show()
```

```python
dane_ad = df.loc["ŚLĄSKIE"]["adiunkci;ogółem;uniwersytety;2014;[osoba]":"adiunkci;ogółem;uniwersytety;2018;[osoba]"]
dane_as = df.loc["ŚLĄSKIE"]["asystenci;ogółem;uniwersytety;2014;[osoba]":"asystenci;ogółem;uniwersytety;2018;[osoba]"]
dane_pr = df.loc["ŚLĄSKIE"]["profesorowie;ogółem;uniwersytety;2014;[osoba]":"profesorowie;ogółem;uniwersytety;2018;[osoba]"]
lata = ["2014", "2015", "2016", "2017", "2018"]

plt.plot(range(len(dane_ad)), dane_ad, "ro-", label="adiunkci")
plt.plot(range(len(dane_as)), dane_as, "bo-", label="asystenci")
plt.plot(range(len(dane_pr)), dane_pr, "yo-", label="profesorowie")
plt.xticks(range(len(lata)), lata)
plt.title('Nauczyciele akademiscy zatrudnieni w latach 2014-2018 w województwie śląskim')
plt.ylabel('osoby', fontsize=15)
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np

dane_ku = np.array(df.loc["ŚLĄSKIE"]["nauczyciele akademiccy;kobiety;uniwersytety;2014;[osoba]":"nauczyciele akademiccy;kobiety;uniwersytety;2018;[osoba]"])
dane_kp = np.array(df.loc["ŚLĄSKIE"]["nauczyciele akademiccy;kobiety;wyższe szkoły techniczne;2014;[osoba]":"nauczyciele akademiccy;kobiety;wyższe szkoły techniczne;2018;[osoba]"])
dane_mu = np.array(df.loc["ŚLĄSKIE"]["nauczyciele akademiccy;ogółem;uniwersytety;2014;[osoba]":"nauczyciele akademiccy;ogółem;uniwersytety;2018;[osoba]"]) - dane_ku 
dane_mp = np.array(df.loc["ŚLĄSKIE"]["nauczyciele akademiccy;ogółem;wyższe szkoły techniczne;2014;[osoba]":"nauczyciele akademiccy;ogółem;wyższe szkoły techniczne;2018;[osoba]"]) - dane_kp

lata = ["2014", "2015", "2016", "2017", "2018"]

plt.plot(range(len(dane_ku)), dane_ku, "ro-", label="kobiety uniwersystety")
plt.plot(range(len(dane_kp)), dane_kp, "bo-", label="kobiety politechniki")
plt.plot(range(len(dane_mu)), dane_mu, "yo-", label="mężczyźni uniwersystety")
plt.plot(range(len(dane_mp)), dane_mp, "go-", label="mężczyźni politechniki")
plt.xticks(range(len(lata)), lata)
plt.title('Nauczyciele akademiscy zatrudnieni w latach 2014-2018 w województwie śląskim')
plt.ylabel('osoby', fontsize=15)
plt.legend(loc='best')
plt.show()
```
